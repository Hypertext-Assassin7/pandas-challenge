```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Files to Load
load_schools_data = "Resources/schools_complete.csv"
load_students_data = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(load_schools_data)
student_data = pd.read_csv(load_students_data)

```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Calculate the total number of schools
school_count = school_data["School ID"].count()
school_count
```

```python
#Calculate the total number of students
student_count = student_data["Student ID"].count()
student_count
```

```python
#Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
```

```python
#Calculate the average maths score
average_math_score = school_data_complete["maths_score"].mean()
average_math_score
```

```python
#Calculate the average reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#Calculate the % of students with a passing maths score (50 or greater)
maths_pass_count = 0
for score in school_data_complete["maths_score"]:
    if score >= 50:
        maths_pass_count += 1
#        maths_pass_count = maths_pass_count + 1

print('%2.6f' , maths_pass_count / student_count)
maths_pass_count / student_count *100
```

```python
#Calculate the % of students with a passing reading score (50 or greater)
reading_pass_count = 0
for score in school_data_complete["reading_score"]:
    if score >= 50:
        reading_pass_count += 1

reading_pass_count/student_count*100
```

```python
#Calculate the % of students who passed maths and reading (% Overall Passing)
double_pass_count = 0
for r_score,m_score in zip(school_data_complete["reading_score"],school_data_complete["maths_score"]):
    if r_score >= 50 and m_score >= 50:
        double_pass_count += 1

double_pass_count/student_count*100
```

```python
#Local government area summary
area_summary = pd.DataFrame({"Total Schools": [school_count],
                              "Total Students": student_count,
                              "Total Budget": total_budget,
                              "Average Maths Score": average_math_score, 
                              "Average Reading Score": average_reading_score,
                              "% Passing Maths": maths_pass_count / student_count *100,
                              "% passing reading": reading_pass_count/student_count*100,
                              "overall Passing": double_pass_count/student_count*100
                            })
area_summary
```

```python
#Create an overview table that summarises key metrics about each school
school_summary = school_data_complete.groupby(["school_name"])
# school_summary[["Student ID"]].count()
result=pd.merge(school_summary[["Student ID"]].count(), school_summary[["maths_score","reading_score"]].mean(), on=["school_name"])
result=pd.merge(result,school_data,on=["school_name"])
print(result)
```

```python
r3=pd.merge(r2,p6,on=["school_name"])
print(r3)
```

```python
p7=school_data_complete.loc[(school_data_complete["reading_score"] >= 50) & (school_data_complete["maths_score"] >= 50), :].groupby(["school_name"]).count()
# pd.merge(result,p1,on=["school_name"])
p8=p7.drop(columns=['student_name',"gender","year","reading_score","maths_score","School ID","type","size","budget"])
p9=p8.rename(columns={"Student ID":"# pass"})
print(p9)
```

```python
r4=pd.merge(r3,p9,on=["school_name"])
print(r4)
```

```python
r4['Per Student Budget']=r4["budget"]/r4["Student ID"]
r4['% Passing Maths']=r4["# math pass"]/r4["Student ID"] * 100
r4['% Passing Reading']=r4["# read pass"]/r4["Student ID"] * 100
r4['% Overall Passing']=r4["# pass"]/r4["Student ID"] * 100
print(r4)
```

```python
#school summary
r5=r4.drop(columns=["size","School ID","# math pass","# read pass","# pass"])
r5a=r5.rename(columns={"Student ID":"Total Students","type":"School Type","maths_score":"Average Maths Score","reading_score":"Average Reading Score",
                     "budget":"Total School Budget"})
r6=r5a.set_index("school_name")
r6.head(550)
```

```python
#Maths Scores by Year
# maths_scores_by_year = student_data.groupby(["school_name"])
# maths_scores_by_year.head()
df_y9=student_data.loc[student_data["year"] == 9, :].groupby(["school_name"])
df_y9ms=df_y9[["maths_score"]].mean()
df_y9ms2=df_y9ms.rename(columns={"maths_score":"Year 9"})
df_y9ms2
df_y10=student_data.loc[student_data["year"] == 10, :].groupby(["school_name"])
df_y10ms=df_y10[["maths_score"]].mean()
df_y10ms2=df_y10ms.rename(columns={"maths_score":"Year 10"})
df_y10ms2
df_y11=student_data.loc[student_data["year"] == 11, :].groupby(["school_name"])
df_y11ms=df_y11[["maths_score"]].mean()
df_y11ms2=df_y11ms.rename(columns={"maths_score":"Year 11"})
df_y11ms2
df_y12=student_data.loc[student_data["year"] == 12, :].groupby(["school_name"])
df_y12ms=df_y12[["maths_score"]].mean()
df_y12ms2=df_y12ms.rename(columns={"maths_score":"Year 12"})
df_y12ms2
df_ms1=pd.merge(df_y9ms2,df_y10ms2,on="school_name")
df_ms2=pd.merge(df_ms1,df_y11ms2,on="school_name")
df_ms3=pd.merge(df_ms2,df_y12ms2,on="school_name")
df_ms3
```

```python
#Reading Score by Year
# reading_scores_by_year = student_data.groupby(["school_name"])
# reading_scores_by_year.head()
df_y9rs=df_y9[["reading_score"]].mean()
df_y9rs2=df_y9rs.rename(columns={"reading_score":"Year 9"})
df_y9rs2
df_y10rs=df_y10[["reading_score"]].mean()
df_y10rs2=df_y10rs.rename(columns={"reading_score":"Year 10"})
df_y10rs2
df_y11rs=df_y11[["reading_score"]].mean()
df_y11rs2=df_y11rs.rename(columns={"reading_score":"Year 11"})
df_y11rs2
df_y12rs=df_y12[["reading_score"]].mean()
df_y12rs2=df_y12rs.rename(columns={"reading_score":"Year 12"})
df_y12rs2
df_rs1=pd.merge(df_y9rs2,df_y10rs2,on="school_name")
df_rs2=pd.merge(df_rs1,df_y11rs2,on="school_name")
df_rs3=pd.merge(df_rs2,df_y12rs2,on="school_name")
df_rs3
```

```python
#Scores by School Spending
bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$580-630", "$630-645", "$645-680"]
r7=r6
r7["Spending Ranges (per Student)"]=pd.cut(r6["Per Student Budget"], bins, labels=group_names)
r8=r7.groupby(["Spending Ranges (per Student)"])
r9=r8[["Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]].mean()
r9["Average Maths Score"]=r9["Average Maths Score"].astype(float).map("{:,.2f}".format)
r9["Average Reading Score"]=r9["Average Reading Score"].astype(float).map("{:,.2f}".format)
r9["% Passing Maths"]=r9["% Passing Maths"].astype(float).map("{:,.2f}".format)
r9["% Passing Reading"]=r9["% Passing Reading"].astype(float).map("{:,.2f}".format)
r9["% Overall Passing"]=r9["% Overall Passing"].astype(float).map("{:,.2f}".format)
r9
```

```python
#Scores by School Size
bins = [0, 999, 1999, 5000]
group_names = ["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]
r10=r6
r10["School Size"]=pd.cut(r6["Total Students"], bins, labels=group_names)
r11=r10.groupby(["School Size"])
r12=r11[["Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]].mean()
r12
```
